```python
import requests
import bs4
url='https://www.amazon.in/dp/8172345666/ref=sspa_dk_detail_3?psc=1&pd_rd_i=8172345666&pd_rd_w=oZucq&content-id=amzn1.sym.2575ab02-73ff-40ca-8d3a-4fbe87c5a28d&pf_rd_p=2575ab02-73ff-40ca-8d3a-4fbe87c5a28d&pf_rd_r=CCX23FR98WHSECRXK9WP&pd_rd_wg=gAgeH&pd_rd_r=8ecf3d48-a0de-4166-8aad-d2f7021a60aa&s=books&sp_csd=d2lkZ2V0TmFtZT1zcF9kZXRhaWw'
Header=({'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36 Edg/124.0.0.0'})
```

```python
request1=requests.get(url,headers=Header)
request1
```

```python
request1.content
soup=bs4.BeautifulSoup(request1.content)
```

```python
reviews=soup.findAll('div',{'class':"a-section a-spacing-large reviews-content filterable-reviews-content celwidget"})
for review in reviews:
    print(review,end="\n")
```

```python
# Fetch customer name
customer_names=soup.find_all('span',{'class' : 'a-profile-name'})
for customer in customer_names:
    print(customer.get_text(),end="\n")
```

```python
tags=soup.find_all('a',{'class' : 'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold'})
for tag in tags:
    print(tag.get_text(),end="\n")
```
